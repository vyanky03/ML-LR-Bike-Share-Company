### 5. Splitting Dataset into train and trest sets

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### 6. Scaling 
- Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

num_vars = ['temp', 'atemp', 'hum','windspeed', 'cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

- Let's check the correlation coefficients to see which variables are highly correlated

#### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

### 7: Feature Selection and Modelling

- I will be using RFE as a part of hybrid model building approach for better and quick generation of model

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

rfe = RFE(lr, n_features_to_select=15)             # running RFE and keeping variable to 10
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

- These are the columns which are qualified by the RFE method

```python
X_train.columns[~rfe.support_]
```

- These columns are not qualified for the further model building process

### 8. Building model with statsmodel

```python
# Checking correlation of features selected by RFE with target column. 
# Also to check impact of different features on target.
plt.figure(figsize = (15,10))
sns.heatmap(df[['yr', 'holiday', 'temp', 'hum', 'windspeed', 'season_2', 'season_3',
       'season_4', 'mnth_3', 'mnth_5', 'mnth_8', 'mnth_9', 'mnth_10',
       'weathersit_2', 'weathersit_3']].corr(), cmap='GnBu', annot=True)
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

#### 8.1 First Model

```python
col=['yr', 'holiday', 'temp', 'hum', 'windspeed', 'season_2', 'season_3',
       'season_4', 'mnth_3', 'mnth_5', 'mnth_8', 'mnth_9', 'mnth_10',
       'weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

#### 8.2 Second Model

```python
# p value of mnth_5 variable is high lets drop it and rebuild the model

col=['yr', 'holiday', 'temp', 'hum', 'windspeed', 'season_2', 'season_3',
       'season_4', 'mnth_3', 'mnth_8', 'mnth_9', 'mnth_10',
       'weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

#### 8.3 Third Model

```python
# p value of season_3 variable is high lets drop it and rebuild the model

col=['yr', 'holiday', 'temp', 'hum', 'windspeed', 'season_2',
       'season_4', 'mnth_3', 'mnth_8', 'mnth_9', 'mnth_10', 'weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

#### 8.4 Fourth Model

```python
# p value of mnth_3 variable is high, lets drop it and rebuild the model

col=['yr', 'holiday', 'temp', 'hum', 'windspeed', 'season_2',
       'season_4', 'mnth_8', 'mnth_9', 'mnth_10', 'weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

- Above model has total 12 features within the recommended range of p value i.e. (0.05)
- The VIF of few columns in well above range i.e. (VIF>5)
- Lets remove few features one by to build model

#### 8.5 Fifth Model

```python
# Removing hum variable as it has high VIF value

col=['yr', 'holiday', 'temp', 'windspeed', 'season_2',
       'season_4', 'mnth_8', 'mnth_9', 'mnth_10', 'weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

- In the above mode, I have removed `hum` feature as it has higher VIF above range
- But, after removing it the p value of `mnth_10` has gone up
- will try and see after removing `mnth_10` what happens to the model

#### 8.6 Sixth Model

```python
col=['yr', 'holiday', 'temp', 'windspeed', 'season_2',
       'season_4', 'mnth_8', 'mnth_9','weathersit_2', 'weathersit_3']

X_train_sm = sm.add_constant(X_train[col])
lr = sm.OLS(y_train, X_train_sm.astype(float)).fit()
print(lr.summary())

X_train_new = X_train_sm.drop(['const'], axis=1)
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.astype(float).values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
print(vif.sort_values(by='VIF',ascending=False))
```

- This model seems to be a good model based on few points as below
1. The p value of all the variable is less than 0.05
2. VIF value of all the variable is well within the range except `temp`
3. But if `temp` removed the r square value is decreasing to 63 from 83 so its not a good approach
- Total 10 features are select based on above model
